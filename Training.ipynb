```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

train = pd.read_csv("train.csv")

test = pd.read_csv("test.csv")
```

#Analisi dei dati

```python
#stampo le feautures
train.columns
```

```python
train.sample(10)
```

```python
train.describe(include='all')
```

```python
train.info()
```

```python
#elimino feautures non necessarie
train.drop(['customerID'], axis =1, inplace=True)
test.drop(['customerID'], axis =1, inplace=True)
```

```python
#Casting feautures 'TotalCharges' from object to float
train['TotalCharges'] = pd.to_numeric(train.TotalCharges, errors='coerce')
test['TotalCharges'] = pd.to_numeric(test.TotalCharges, errors='coerce')

#train['tenure'] = train['tenure'].astype(float)
#test['tenure'] = test['tenure'].astype(float)
```

```python
train['Churn'].replace(to_replace='Yes', value=1, inplace=True)
train['Churn'].replace(to_replace='No',  value=0, inplace=True)

test['Churn'].replace(to_replace='Yes', value=1, inplace=True)
test['Churn'].replace(to_replace='No',  value=0, inplace=True)
```

```python
train.info()
```

```python
#controllo valori nulli, ci sono 9 valori nulli di TotalCharges nel train e 1 nel test
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
media = train['TotalCharges'].mean()
train['TotalCharges'] = train['TotalCharges'].replace(np.NaN, media).astype('float64')
media = test['TotalCharges'].mean()
test['TotalCharges'] = test['TotalCharges'].replace(np.NaN, media).astype('float64')
```

```python
train.isnull().sum()
```

```python
s =(train.dtypes == 'object')
object_cols = list(s[s].index)
print(object_cols)
train2 = train.drop(object_cols, axis =1)
train3 = pd.get_dummies(train[object_cols])
train = pd.concat([train3,train2], axis = 1)
train
```

```python
s =(test.dtypes == 'object')
object_cols = list(s[s].index)
print(object_cols)
test2 = test.drop(object_cols, axis =1)
test3 = pd.get_dummies(test[object_cols])
test = pd.concat([test3,test2], axis = 1)
test
```

#Addestramento del modello

```python
y_train = train["Churn"].values #valori di class
X_train = train.drop(["Churn"], axis=1).values #tutti i dati eccetto la colonna "class"

y_test = test["Churn"].values #valori di class
X_test = test.drop(["Churn"], axis=1).values #tutti i dati eccetto la colonna "class"
```

```python
from sklearn.linear_model import LogisticRegression

logic = LogisticRegression(random_state=0, max_iter=1000)
logic.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(logic.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(logic.score(X_test, y_test)))
```
